```python
# import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read the csv file 
calgary_listing = pd.read_csv('data/calgary_listings-test.csv')
score_df = pd.read_csv('data/score_df.csv')
```

```python
Data = pd.merge(calgary_listing, score_df, on='postal_code')
Data.head()
```

```python
Data.info()
Data.describe().transpose()
```

```python
Data.isnull().sum()
```

data_not_null = Data.dropna(subset=['name', 'born'])
data_not_null

```python
prop_area = pd.get_dummies(Data["property_type"])
prop_area.head()
```

```python
table_df = pd.concat([Data, prop_area], axis=1)
```

```python
data_raw = table_df.filter(['price', 'bed', 'full_bath', 'half_bath', 'property_area', 'years_old', 'distance_downtown',
                'monthly_payments', 'lot_size', 'basement', 'garage', 'walk_score', 'bike_score', 'transit_score','house', 'condo', 'townhouse'], axis=1)
data_raw.head()
```

```python
data_raw['lot_size'] = data_raw['lot_size'].replace('na', np.nan)
data_raw['half_bath'] = data_raw['half_bath'].replace(np.nan, 0)
data_raw.head()
```

```python
len(data_raw)
```

```python
data_raw_na = data_raw.dropna()
```

```python
len(data_raw_na)
```

```python
data_raw_na.isnull().sum()
```

```python
X = data_raw_na.drop('price',axis =1).values
y = data_raw_na['price'].values

#splitting Train and Test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=12)
```

```python
#standardization scaler - fit&transform on train, fit only on test
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
X_train = s_scaler.fit_transform(X_train.astype(np.float))
X_test = s_scaler.transform(X_test.astype(np.float))
```

```python
#Liner Regression
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train)

#evaluate the model (intercept and slope)
regressor.intercept_
regressor.coef_
```

```python
y_predd = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predd})
df1 = df.head(10)
df1
```

```python
fig = plt.figure(figsize=(10,5))
residuals = (y_test- y_predd)
sns.distplot(residuals)
```

```python
from sklearn import metrics

print('Mean Absolute Error: {:.2f}'.format(metrics.mean_absolute_error(y_test, y_predd))) 
print('Mean Squared Error:{:.2f}'.format(metrics.mean_squared_error(y_test, y_predd)))  
print('Root Mean Squared Error:{:.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, y_predd))))
print('Variance score is: {:.2f}'.format(metrics.explained_variance_score(y_test,y_predd)))
```

```python
# we are off about 20% (comparing mean absolut error and mean of price)
Data['price'].mean()
```

```python
print('Linear Regression Model:')
print("Train Score {:.2f}".format(regressor.score(X_train,y_train)))
print("Test Score {:.2f}".format(regressor.score(X_test, y_test)))
```

```python
# Creating a Neural Network Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
```

```python
# having 19 nueron is based on the number of available featurs

model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
          validation_data=(X_test,y_test),
          batch_size=128,epochs=400)
```
